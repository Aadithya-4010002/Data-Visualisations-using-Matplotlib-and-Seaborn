```python
def mean(values):
    return sum(values)/float(len(values))
```

```python
def variance(values,mean):
    return sum([(x-mean)**2 for x in values])
```

```python
def covariance(x,mean_x,y,mean_y):
    covar=0.0
    for i in range(len(x)):
        covar+=(x[i]-mean_x) * (y[i]-mean_y)
    return covar
```

```python
def coefficients(dataset):
    b1=covariance(x,mean_x,y,mean_y)/variance(x,mean_x)
    b0=mean_y-b1*mean_x
    return[b0,b1]
```

```python
def simple_linear_regression(train,test):
    for row in test:
        ytest = b0 + b1 * row[0]
    return ytest
```

```python
dataset=[[30,18],[40,70],[18,45],[60,70],[15,30],[90,68],[75,90]]
x=[row[0] for row in dataset]
y=[row[1] for row in dataset]
mean_x=mean(x)
mean_y=mean(y)
```

```python
variance_x=variance(x,mean_x)
variance_y=variance(y,mean_y)
print('x stats:mean=%.3f variance=%.3f' % (mean_x,variance_x))
print('y stats:mean=%.3f variance=%.3f' % (mean_y,variance_y))
```

```python
covar = covariance(x,mean_x,y,mean_y)
print('covariance: %.3f' % (covar))
```

```python
b0,b1 = coefficients(dataset)
print('coefficients:b0=%.3f,b1=%.3f' % (b0,b1))
print('Regression equation of y on x : y=%.3f+%.3fx '% (b0, b1))
```

```python
import numpy as np
from sklearn.linear_model import LogisticRegression
import matplotlib.pyplot as plt

dataset = np.array([[30, 0], [40, 1], [18, 0], [60, 1], [15, 0], [90, 1], [75, 1]])

x = dataset[:, 0].reshape(-1, 1)
y = dataset[:, 1]

logistic_model = LogisticRegression()
logistic_model.fit(x, y)

predicted_probs = logistic_model.predict_proba(x)
predicted_labels = logistic_model.predict(x)

print("Predicted Probabilities:")
print(predicted_probs)

print("Predicted Class Labels:")
print(predicted_labels)

x_range = np.linspace(0, 100, 100).reshape(-1, 1)
y_prob = logistic_model.predict_proba(x_range)[:, 1]

plt.figure(figsize=(10, 6))
plt.scatter(x, y, c=y, cmap=plt.cm.coolwarm, marker='o', edgecolors='k')
plt.plot(x_range, y_prob, color='blue', linewidth=2)
plt.xlabel("Popularity")
plt.ylabel("Class (0 or 1)")
plt.title("Logistic Regression Decision Boundary")
plt.show()
```
